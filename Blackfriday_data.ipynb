```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#plt.rcParams['figure.figsize'] = (12, 6)
```

```python
train=pd.read_csv("C:/Users/a3dullahi/Dropbox/PC/Desktop/blackfriday/train.csv")
train
```

```python
test = pd.read_csv("C:/Users/a3dullahi/Dropbox/PC/Desktop/blackfriday/test.csv")
test
```

```python
df = pd.concat([train, test], ignore_index=True)
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(['User_ID'], axis = 1, inplace = True)
df
```

```python
#Handling categorical features Gender
df['Gender']=df['Gender'].map({'F':0, 'M':1})
df.head()
```

```python
df['Age'].unique()
```

```python
#Handling categorical features Age
#pd.get_dummies(df['Age'],drop_first=True)
df['Age'] = df['Age'].map({'0-17':1,'18-25':2 ,'26-35':3, '36-45':4, '46-50':5, '51-55':6, '55+':7})
df.head()
```

```python
df['City_Category'].unique()
```

```python
 # Import label encoder 
from sklearn import preprocessing 
  
# label_encoder object knows  
# how to understand word labels. 
label_encoder = preprocessing.LabelEncoder() 
  
# Encode labels in column 'City_Category'. 
df['City_Category']= label_encoder.fit_transform(df['City_Category']) 
  
df['City_Category'].unique()
```

```python
#pinpointing columns with missing values
[factor for factor in df.columns if df[factor].isnull().sum()>0]
```

```python
#
df.isnull().sum()/df.shape[0]*100
```

```python
#This is a discrete feature. It's best to replace null values with the mode.
df['Product_Category_2'].unique()
```

```python
df['Product_Category_2'].mode()
```

```python
# Fill missing values in 'Product_Category_2' column with the mode
df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0], inplace=True)
```

```python
#This is also a discrete feature.
df['Product_Category_3'].unique()
df['Product_Category_3'].mode()
# Fill missing values in 'Product_Category_3' column with the mode
df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0], inplace=True)
```

```python
df.head()
```

```python
# Replace '+' character with an empty string
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].str.replace('+', '')
df.head()
```

```python
# Convert 'Stay_In_Current_City_Years' column to int type
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].astype(int)
df.info()
```

```python
#sns.pairplot(df)
```

```python
# Plot the barplot
sns.barplot(x='Age', y='Purchase', hue='Gender', data=df)

# Add a title
plt.title("Average Purchase")

# Add labels to the axes
plt.xlabel("Age")
plt.ylabel("Average Purchase Amount")

# Show the plot
plt.show()
```

### According to the analysis of purchase data, it appears that males tend to buy more items compared to females, regardless of age. Additionally, age does not appear to significantly influence purchase behavior, as the average purchase amount remains consistent across different age groups within each gender.
